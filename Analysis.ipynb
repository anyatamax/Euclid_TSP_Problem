## Анализ алгоритма и времени работы

```python
import numpy as np
import matplotlib.pyplot as plt
```

Анализ был проведен с помощью генерации рандомных полных метрических графов и сравнения алгоритма полного перебора и алгоритма Кристофидеса. Ответы при тестировании первого алгоритма были записаны в файлы answers1.txt и time1.txt, а при полном переборе в файлы answers2.txt и time2.txt

Сначала загрузим все данные в массив и построим график отношения ответов, которые дал алгоритм Кристофидеса, к правильным ответам, полученным с помощью перебора

```python
algo_ans = np.loadtxt('answers1.txt')
cur_ans = np.loadtxt('answers2.txt')

plt.figure(figsize=(10, 10))
plt.plot(np.arange(3, 14), algo_ans / cur_ans, label='Отношение ответов алгоритма к правильным')
plt.legend()
plt.title('Отношение ответов алгоритма Кристофидеса к правильным ответам')
plt.ylim(1, 1.5)
plt.xlabel('Количество вершин')
plt.show()
```

Далее посмотрим на время, затраченное двумя алгоритмами

```python
algo_time = np.loadtxt('time1.txt')
cur_time = np.loadtxt('time2.txt')

plt.figure(figsize=(10, 10))
plt.plot(np.arange(3, 14), algo_time, label='Алгоритм Кристофидеса')
plt.plot(np.arange(3, 14), cur_time, label='Алгоритм перебора')
plt.legend()
plt.title('Время работы алгоритмов')
plt.xlabel('Количество вершин')
plt.show()
```

Проведя небольшое исследование, было выяснено, что алгоритм Кристофидеса действительно дает 1.5-апроксимацию, то есть ответ алгоритма отличается не более чем в 1.5 раза от правильного. Но зато алгоритм Кристофидеса является намного быстрее обычного перебора, о чем говорит нам последний график. Еще можно заметить, что ответ отличается от правильно начиная с 10 вершин. То есть исследуемый алгоритм дает почти всегда правильные ответы при маленьком количестве вершин. Класс графов, которые можно исследовать ограничен полными метрическими графами, поэтому данного тестирования достаточно, чтобы сделать адекватные выводы о работе алгоритма Кристофидеса.
